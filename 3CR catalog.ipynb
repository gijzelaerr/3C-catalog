```python
file_ = 'original/3cr.dat'
```

```python
def extract(line, start, stop):
    return line[start:stop].strip()


def dms2deg(hour, min_, sec):
    deg = abs(hour) * 15 # there are 15 degrees in one hour
    deg += min_ / 4 # there are 4 minutes in one degree
    deg += sec / 240 # there are 240 seconds in one degree

    if hour < 0:
        return deg*-1
    return deg
    
def sdm2deg(sign, deg, min_):
    deg = abs(deg)
    deg += min_ / 60
    
    if sign == "-":
        deg = deg * -1
    return deg
    
   
```

```python
catalog = []

for line in open(file_, 'r').readlines():
    id_ = extract(line, 0, 5)                 #  3CR source designation (1)
    RAh = int(extract(line, 7, 9))            #  Right ascension in h (1950.0) 
    RAm = int(extract(line, 10, 12))          #  Right ascension in min (1950.0)
    RAs = float(extract(line, 13, 17))        #  Right ascension in s (1950.0)
    e_RAs = float(extract(line, 19, 23))      #  rms uncertainty on right ascension ins
    
    ra = dms2deg(RAh, RAm, RAs)
    
    sign = extract(line, 24, 25)              #  Declination sign (1950.0)
    DEd = float(extract(line, 25, 27))        #  Declination (1950.0) in deg
    DEm = float(extract(line, 28, 32))        #  Declination (1950.0) in arcmin 
    
    decl = sdm2deg(sign, DEd, DEm)

    e_DEm = float(extract(line, 33, 39))      #  rms uncertainty on declination in arcmin
    GLON = float(extract(line, 40, 43))       #  Galactic longitude in deg
    GLAT = float(extract(line, 45, 48))       #  Galactic latitude in deg
    flux = float(extract(line, 49, 56))       #  Flux density at 178MHz in Jy
    flux_  = extract(line, 56, 57)            #  469{lambda} interferometer (2)
    l_Diam = extract(line, 60, 61)            #  Limit flag on Diam
    Diam = extract(line, 61, 65)              #  Angular diameter
    x_Diam = extract(line, 65, 66)            #  ["'] Units of diameter (' or ")
    e_Diam = extract(line, 66, 70)            #  rms uncertainty on Diam
    f_Diam =extract(line, 70, 71)             #  [)] Confusing sources
    notes = extract(line, 72, 208)            #  Notes
    
    catalog.append({'ra': ra, 'decl': decl, 'flux':flux, 'id': id_})

print catalog

    
```
